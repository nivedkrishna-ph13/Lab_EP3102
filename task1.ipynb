```python
import time
import csv
import pandas as pd
name = input("enter your name: ")
n = -1
time_begin=[]
time_end=[]
value_=[]
delay_=[]
while(1):
    x = time.time()
    n = input("enter a number between 0-9(type 'quit' if you want to exit):")
    y= time.time()
    delay = y-x
    if (n=='quit'):
        break
    
    n = int(n);
    if n<-1 or n>10:
        print('please enter a valid number')
        continue
    else:
        time_begin.append(x)
        time_end.append(y)
        value_.append(n)
        delay_.append(delay)
    
                    
rows = zip(value_, time_begin,delay_)

headers = ['value','timestamp','delay_']

# Specify the output CSV filename
output_filename = 'output_data.csv'

# Write the data to the CSV file
with open(output_filename, 'w', newline='') as csvfile:
    csv_writer = csv.writer(csvfile)

    # Write the header row
    csv_writer.writerow(headers)

    # Write the data rows
    for row in rows:
        csv_writer.writerow(row)

print(f"Data successfully written to {output_filename}")
    
```
